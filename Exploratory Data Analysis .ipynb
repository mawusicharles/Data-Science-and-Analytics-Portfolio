```python
import pandas as pd
import numpy as np
```

```python
path='https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/automobileEDA.csv'
df = pd.read_csv(path)
df.head()
```

### Preliminary Data Visualization 
<a id="pattern_visualization"> Analyzing Individual Feature Patterns using Visualization</a>

```python
pip install seaborn
```

```python
# import data visualization packages 
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
# checking for data types 
print(df.dtypes)
```

```python
# correlation matrix for the whole dataframe 
corr= df.corr()
corr
```

```python
# Using seaborn to plot correlation heatmap for the whole data

sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns)
```

```python
# Exploring the correlation between bore, stroke,compression-ratio , and horsepower.

df_corr = df[["bore", "stroke","compression-ratio",  "horsepower"]].corr()
df_corr
```

```python
sns.heatmap(df_corr, xticklabels=df_corr.columns, yticklabels=df_corr.col)
```
