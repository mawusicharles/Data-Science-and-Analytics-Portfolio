## Data Extraction and Preparation 

I Extracted the data using SQL. The syntax used to retrived the data includes, select city,avg_temp from city_data  where city ='Paris' ; select* from global_data; select*  from city_list; respectively.
 
I then merged tables and calculated the 5 year moving average and 10 year moving averages respectively using excel. The goal is smooth out the yearly fluctuations and make it easier to observe long term trends. 

### Calculating the Moving Averages 

In calculating the 5 year moving average for Paris (AVGF_Paris, also corresponds to column D in Excel) i took mean of the first five rows of the columne labelled paris(also corresponds to column B in excel). The excel syntax looks like this AVERAGE(B2:B6) which returns a value of 9.95. I clicked and dragged this cell to the bottom to compute the moving averages for the remainder of values in column B (Paris). The 5 year moving average is thus, what we see in the column labelled AVGF_Paris. I took similar approch in calculating the 10 year moving average.

## Comparing the two trends (Global vs Paris)

```python
plt.plot(year, GlobeTen, label ="Globe")
plt.plot(year, parisTen, label ="Paris")
plt.title("Graph Temperature-10 year moving averages ")
plt.xlabel("Year")
plt.ylabel("Temperatures")
plt.legend()
plt.show()
```

## Observations from the Graph 
Temperatures in paris and the rest of the world seem to follow similar tragectory. However, the temperatures in paris has been consistently higher than the global average. Athough the average temperatures seems to fluctuate more in paris than the globe, the overall trend seem to suggest that temperatures in both cases has been rising. This rise is particulalry pronounced in the  the 20th century (1900-2000). 

Finally i observe that wheras the average global temperures seems to continue with its upward trend in recent times (after year 2000) temperatures in paris seem to follow a downwards trend.
